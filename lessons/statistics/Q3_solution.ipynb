**Exercise:** The numbers generated by `numpy.random.random` are supposed to be uniform between 0 and 1; that is, every value in the range should have the same probability.

Generate 1000 numbers from `numpy.random.random` and plot their PMF.  What goes wrong?

Now plot the CDF. Is the distribution uniform?

```python
sample = np.random.random(1000)
sample_pmf = thinkstats2.Pmf(sample)
thinkplot.Pmf(sample_pmf, linewidth = 0.1)
thinkplot.Config(xlabel='Number between 0 and 1', ylabel='PMF')
```

It is difficult to tell if every value truly has the same probability, but since the CDF evalutes probabilty for any value in the range of the sample, we should be able to look at the probality overall, not biased by the values we happened to get in our random sample.

```python
sample_cdf = thinkstats2.Cdf(sample, label = "random numbers")
thinkplot.Cdf(sample_cdf)
thinkplot.Config(xlabel='Number between 0 and 1', ylabel='CDF')
```
